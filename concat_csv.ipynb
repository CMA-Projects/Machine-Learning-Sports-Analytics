```python
import pandas as pd
import os

# Define the folder path
folder_path = '../Sports-Analytics-Machine-Learning/data/Golf Stats PGA Tour'

# List all the CSV files
csv_files = [files for files in os.listdir(folder_path) if files.endswith('.csv')]

# Create an empty lists to store Dataframes
dfs = []

# Loop through the list of CSV files and read each one into a df
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dfs.append(df)

# Combine all the dtaframes inside the dfs list into one
combined_df = pd.concat(dfs, ignore_index = True, sort = False)
```

```python
# Set options to display more rows and columns
pd.set_option('display.max_rows', None)  # Display all rows
pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.max_colwidth', None)  # Display full width of each column

print(combined_df)
```

```python
combined_df.info()
```

```python
aggregated_df = combined_df.groupby('PLAYER_ID', as_index = False).first()

print(aggregated_df)
```

```python
aggregated_df.to_csv('aggregated_df.csv', index = False)
```
